```python
import pandas as pd
```

```python
data1 = pd.read_csv('/content/drive/MyDrive/SemEval Task3/SemEval2024_task3_train.csv' )  #SemEval2024_task3_train.csv
data1= data1[["text", "emotion"]]
data1
```

```python
data2 = pd.read_csv('/content/drive/MyDrive/SemEval Task10/MaSaC_train_erc.csv' )   #MaSaC_train_erc.csv
data2.rename(columns={"utterances_English":"text", "emotions": "emotion"}, inplace = True)
data2 = data2[["text", "emotion"]]
data2
```

```python
nidaty = pd.read_csv('/content/drive/MyDrive/SemEval Task3/Nikam_emotion_dataset.csv' )   #Nikam_emotion_dataset.csv
nidaty.drop(["Unnamed: 0", "Text"], axis=1, inplace = True)
nidaty.rename(columns = {"Clean_Text":"text", "Emotion":"emotion"}, inplace=True)
nidaty.reindex(columns=["text", "emotion"])
nidaty.dropna(inplace=True)
```

[nidata by SannketNikam (anger, disgust, fear, joy, neutral, sadness, shame and surprise)](https://github.com/SannketNikam/Emotion-Detection-in-Text)

```python
nidata = pd.read_csv('/content/drive/MyDrive/SemEval Task3/Nikam_emotion_dataset.csv' )   #Nikam_emotion_dataset.csv
nidata.drop(["Unnamed: 0", "Text"], axis=1, inplace = True)
nidata.rename(columns = {"Clean_Text":"text", "Emotion":"emotion"}, inplace=True)
nidata.reindex(columns=["text", "emotion"])
nidata.dropna(inplace=True)
nidata.drop(nidata.index[nidata.emotion=="shame"], inplace=True)
nidata.set_axis([i for i in range(0, len(nidata))], axis = 0, inplace = True)
```

```python
nidata   #neutral:1811   surprise:4062   anger:4286   sadness:6719  joy:11037  disgust:856  fear:5409
```

```python
nidata.emotion.unique()
```

```python
data1_neutral = pd.concat([data1[data1.emotion=="neutral"], data2[data2.emotion=="neutral"]], axis = 0)
data1_surprise = pd.concat([data1[data1.emotion=="surprise"], data2[data2.emotion=="surprise"], nidata[nidata.emotion=="surprise"]], axis = 0)
data1_anger = pd.concat([data1[data1.emotion=="anger"], data2[data2.emotion=="anger"], nidata[nidata.emotion=="anger"]], axis = 0)
data1_sadness = pd.concat([data1[data1.emotion=="sadness"], data2[data2.emotion=="sadness"], nidata[nidata.emotion=="sadness"]], axis = 0)
data1_joy = pd.concat([data1[data1.emotion=="joy"], data2[data2.emotion=="joy"], nidata[nidata.emotion=="joy"]], axis = 0)
data1_disgust =  pd.concat([data1[data1.emotion=="disgust"], data2[data2.emotion=="disgust"], nidata[nidata.emotion=="disgust"]], axis = 0)
data1_fear = pd.concat([data1[data1.emotion=="fear"], data2[data2.emotion=="fear"], nidata[nidata.emotion=="fear"]], axis = 0)
```

```python
data1_disgust
```

```python
data1_neutral.set_axis([i for i in range(0, len(data1_neutral))], axis = 0, inplace = True)   #370 because previous notebooks had 370 emotion size for each emotion
data1_neutral370 = data1_neutral[data1_neutral.index <= 7000]

data1_surprise.set_axis([i for i in range(0, len(data1_surprise))], axis = 0, inplace = True)
data1_surprise370 = data1_surprise[data1_surprise.index <= 5000]

data1_anger.set_axis([i for i in range(0, len(data1_anger))], axis = 0, inplace = True)
data1_anger370 = data1_anger[data1_anger.index <= 5000]

data1_sadness.set_axis([i for i in range(0, len(data1_sadness))], axis = 0, inplace = True)
data1_sadness370 = data1_sadness[data1_sadness.index <= 5000]

data1_joy.set_axis([i for i in range(0, len(data1_joy))], axis = 0, inplace = True)
data1_joy370 = data1_joy[data1_joy.index <= 5000]

data1_disgust.set_axis([i for i in range(0, len(data1_disgust))], axis = 0, inplace = True)
data1_disgust370 = data1_disgust[data1_disgust.index <= 5000]

data1_fear.set_axis([i for i in range(0, len(data1_fear))], axis = 0, inplace = True)
data1_fear370 = data1_fear[data1_fear.index <= 5000]
```

```python
data1_disgust370
```

```python
data1_370 = pd.concat([data1_neutral370,data1_surprise370, data1_anger370, data1_sadness370, data1_joy370, data1_disgust370, data1_fear370])
```
