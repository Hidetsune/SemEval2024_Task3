```python
json_open = open("/content/drive/MyDrive/SemEval Task10/MaSaC_val_erc.json","r")  #MaSaC_val_erc.json
json_load = json.load(json_open)
```

```python
train_data = json_load
train_data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = ["episode", "utterances", "speakers", "emotions"])
```

```python
df
```

```python
episode_list = []
utterance_list = []
speakers_list = []
emotions_list = []

for episode in train_data:
  if len(episode["speakers"]) == len(episode["utterances"]):
    for i in range(len(episode["speakers"])): episode_list.append(episode["episode"])
    for utterence in episode["utterances"]: utterance_list.append(utterence)
    for speaker in episode["speakers"]: speakers_list.append(speaker)
    for emotion in episode["emotions"]: emotions_list.append(emotion)
  else:
    print("myerror")
```

```python
len(emotions_list)
```

```python
df["episode"] = episode_list
df["utterances"] = utterance_list
df["speakers"] = speakers_list
df["emotions"] = emotions_list
```
